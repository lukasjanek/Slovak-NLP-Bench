```python
import torch
from transformers import T5ForConditionalGeneration, T5Tokenizer
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "gaussalgo/mt5-base-generative-QA_en-cs"

print(f"Loading {model_name}...")

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Move to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

def get_answer(question: str, context: str, model, tokenizer, device):
    input_text = f"question: {question} context: {context}"
    inputs = tokenizer.encode(
        input_text,
        return_tensors="pt",
        truncation=True,
        max_length=512
    ).to(device)

    output_ids = model.generate(
        inputs,
        max_length=64,
        num_beams=5,
        early_stopping=True,
        no_repeat_ngram_size=3
    )

    answer = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return answer

question = "Aké hlavné politické a hospodárske dôvody viedli k rozdeleniu Českej a Slovenskej Federatívnej Republiky, a aké boli okamžité dôsledky tohto rozdelenia na medzinárodné postavenie nových štátov?"
context = "Rozdelenie Českej a Slovenskej Federatívnej Republiky na dva samostatné štáty, Českú republiku a Slovenskú republiku, ktoré sa uskutočnilo 1. januára 1993, bolo výsledkom komplexného vývoja po páde komunistického režimu v roku 1989. Kľúčové politické dôvody zahŕňali odlišné predstavy českých a slovenských politických elít o usporiadaní štátu – kým česká strana preferovala silnú federáciu s centralizovanou vládou, slovenská strana sa zasadzovala za voľnejšiu konfederáciu, ktorá by zabezpečila väčšiu suverenitu pre Slovensko. Hospodárske faktory tiež zohrávali významnú úlohu; Slovensko sa obávalo, že jeho ekonomika, ktorá bola tradične viac závislá od ťažkého priemyslu a zbrojárskej výroby, bude v rámci federácie v nevýhodnej pozícii voči vyspelejšej českej ekonomike. Proces rozdelenia bol dohodnutý politickými predstaviteľmi, najmä Václavom Klausom za českú stranu a Vladimírom Mečiarom za slovenskú. Okamžitými dôsledkami na medzinárodné postavenie bolo, že oba štáty museli nanovo budovať svoje diplomatické vzťahy a získať medzinárodné uznanie. Obidva štáty sa stali členmi Organizácie Spojených národov a začali sa orientovať na vstup do Európskej únie a NATO, pričom tento proces trval niekoľko rokov."

predicted_answer = get_answer(question, context, model, tokenizer, device)

print(f"\nQuestion: {question}")
print(f"Predicted Answer: {predicted_answer}")
```
